Creating the Decision tree classifier and visualize it.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.tree import DecisionTreeRegressor
print('Library import success')
```

```python
df = pd.read_csv("Iris.csv")
df
```

```python
df.head()
```

```python
df.info()
df.describe()
```

```python
df.Species.value_counts()
```

From the above result we can see that there are no missing values.

```python
X = df.iloc[:,1:-1]
Y = df.iloc[:,-1]
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.1, random_state = 0)
```

```python
df = df.iloc[:,1:]
sns.pairplot(df, hue = 'Species')
```

We can see the difference that one species-IRIS SETOSA has different attributes from the other 2 species

```python
lm = DecisionTreeClassifier(random_state = 0)
lm.fit(X_train,Y_train)
```

```python
y_train_pred = lm.predict(X_train)
y_test_pred = lm.predict(X_test)
print("Train Accuracy Score = ", accuracy_score(Y_train,y_train_pred))
print("Test Accuracy Score = ", accuracy_score(Y_test,y_test_pred))
```

```python
plt.figure(figsize = (15,15))
plot_tree(lm, filled = True)
print()
```
